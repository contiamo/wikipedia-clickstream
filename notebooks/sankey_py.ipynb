```python
dat = pd.read_parquet('/labs/data/processed/wikipedia_referer.parq')
```

```python
df = dat[(dat['prev'] == 'London') | (dat['curr'] == 'London')]
```

## Plotly

```python
node = dict(
    pad=15,
    thickness=15,
    line=dict(color='black', width=0.5),
)

link = dict(
    source=df['prev'],
    target=df['curr'],
    value=df['n'],
)

data_trace = dict(
    type='sankey',
    domain=dict(x=[0, 1], y=[0, 1]),
    orientation='h',
    valueformat='.0f',
    valuesuffix='TWh',
    node=node,
    link=link,
)

layout = dict(
    title='London Wikipedia Clickstream',
    font=dict(size=10)
)

fig = dict(data=[data_trace], layout=layout)
py.iplot(fig, validate=False)
```
